```python
from tkinter import *
from tkinter import ttk
from tkinter import filedialog
from pytube import YouTube


folder_name=""

def openLocation():
    global folder_name
    folder_name = filedialog.askdirectory()
    if(len(folder_name)>1):
        locationerror.config(text=folder_name, fg="blue")
    
    else:
        locationerror.config(text="Please Choose valid Folder", fg="red")


def downloadvideo():
    choice=choices.get()
    url=Entry.get()
    
    if(len(url)>1):
        errormessage.config(text="")
        yt=YouTube(url)
        
        if(choice == choices[0]):
            select = yt.streams.filter(progressive=True).first()

        elif(choice == choices[1]):
            select = yt.streams.filter(progressive=True,file_extension='mp4').last()

        elif(choice == choices[2]):
            select = yt.streams.filter(only_audio=True).first()

        else:
            ytdError.config(text="Not a valid link",fg="red")
    
    select.download(folder_name)
    errormessage.config(text="Download Completed")

        


root=Tk()
root.title("YouTube Downloader")
root.geometry("400x400")
root.columnconfigure(0,weight=1)

mainLabel = Label(root, text="Enter URL of the Video", font=("calibri", 16))
mainLabel.grid()

EntryVar = StringVar()
Entry=Entry(root, width=60, textvariable=EntryVar)
Entry.grid()

errormessage = Label(root, text="Link not valid", fg="red",font=("calibri",12))


savelabel = Label(root, text="Save the video file", font=("calibri",15,"bold"))
savelabel.grid()

saveentry=Button(root, width =10, bg ="blue", fg="white", text="Select File Path")
saveentry.grid()

locationerror=Label(root, text="Path Error", fg="red", font=("calibri", 12))


quality=Label(root, text="Select Quality", font=("calibri",16))
quality.grid()

choices=("HD","SD","Audio Only")
choices =ttk.Combobox(root,values=choices)
choices.grid()

downloadbutton=Button(root,width=10, text="DOWNLOAD", bg="green", fg="black",font=("calibri",12))
downloadbutton.grid()



root.mainloop()
```
